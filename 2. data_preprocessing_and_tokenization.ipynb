```python
from tqdm.notebook import tqdm
import re
from konlpy.tag import Komoran


tqdm.pandas()

komoran = Komoran(userdic = "./data/user.dic")
def clean_text(text):# 이모티콘제거
    if isinstance(text, str):
        text = re.sub(r'[^\x00-\x7F\uAC00-\uD7AF]', '', text)
        text = re.sub(r'[!]', '', text)
        return text
    else:
        return ''  

def process_review(sentence):# 토큰화 진행(명사가져오기, outword_list의 단어 빼기)
    try:
        clean_sentence = clean_text(sentence) 
        tokens = komoran.pos(clean_sentence)
        nouns = [x[0] for x in tokens if x[1] in ("NNG", "NNP") and x[0] not in suncream_outword]
        return nouns
    except UnicodeDecodeError as e:
        print(f"UnicodeDecodeError processing sentence: {sentence}")
        print(e)
        return []
    except Exception as e:
        print(f"Error processing sentence: {sentence}")
        print(e)
        return []


suncream_review_df['processed_review'] = suncream_df['review'].progress_apply(process_review)
```

```python
#파일저장
suncream_review_df.to_csv("./preprocessed_data/preprocessed_suncream_data.csv",encoding="utf-8-sig",index=False)
```

```python
from tqdm.notebook import tqdm
import re
from konlpy.tag import Komoran


tqdm.pandas()

komoran = Komoran(userdic = "./data/user.dic")
def clean_text(text):# 이모티콘제거
    if isinstance(text, str):
        text = re.sub(r'[^\x00-\x7F\uAC00-\uD7AF]', '', text)
        text = re.sub(r'[!]', '', text)
        return text
    else:
        return ''  

def process_review(sentence):# 토큰화 진행(명사가져오기, outword_list의 단어 빼기)
    try:
        clean_sentence = clean_text(sentence) 
        tokens = komoran.pos(clean_sentence)
        nouns = [x[0] for x in tokens if x[1] in ("NNG", "NNP") and x[0] not in sunspray_outword]
        return nouns
    except UnicodeDecodeError as e:
        print(f"UnicodeDecodeError processing sentence: {sentence}")
        print(e)
        return []
    except Exception as e:
        print(f"Error processing sentence: {sentence}")
        print(e)
        return []


sunspray_reviews_df['processed_review'] = sunspray_reviews_df['review'].progress_apply(process_review)
```
