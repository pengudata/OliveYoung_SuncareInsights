# 3-1. 선케어 제품 전체 카테고리 EDA

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

plt.rcParams['font.family'] = 'NanumGothic'
```

## 카테고리별 제품 개수

```python
# 카테고리별 파일 링크
product_count_file_paths = {"선크림": "./crawling_data/suncream_link.csv",
              "선스틱": "./crawling_data/sunstic_link.csv",
              "선쿠션": "./crawling_data/suncushion_list.csv",
              "선패치": "./crawling_data/sunpatch_link.csv",
              "선스프레이": "./crawling_data/sunspray_link.csv"}
```

```python
product_counts = {}
```

```python
# 카테고리별 제품개수 구하기
for category, file_path in product_count_file_paths.items():
    df = pd.read_csv(file_path)
    product_counts[category] = len(df)
```

```python
categories = list(product_counts.keys())
counts = list(product_counts.values())
plt.figure(figsize=(10, 6))
bars = plt.bar(categories, counts, color='skyblue')
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 5, int(yval), ha='center', va='bottom')
plt.title("카테고리별 제품 개수")
plt.xlabel("카테고리")
plt.ylabel("수")
plt.tight_layout()
plt.show()
```

## 카테고리별 리뷰 개수

```python
# 카테고리별 리뷰 데이터 링크
review_count_file_paths = {"선크림": "./crawling_data/suncream_reviews.csv",
              "선스틱": "./crawling_data/sunstick_reviews.csv",
              "선쿠션": "./crawling_data/suncushion_reviews.csv",
              "선패치": "./crawling_data/sunpatch_reviews&buyrank.csv",
              "선스프레이": "./crawling_data/sunspray_reviews&buyrank.csv"}
```

```python
review_counts = {}
```

```python
# 카테고리별 리뷰 개수 구하기
for category, file_path in review_count_file_paths.items():
    df = pd.read_csv(file_path)
    review_counts[category] = len(df)
```

```python
categories = list(review_counts.keys())
counts = list(review_counts.values())
plt.figure(figsize=(10, 6))
bars = plt.bar(categories, counts, color='darkblue')
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 5, int(yval), ha='center', va='bottom')
plt.title("카테고리별 리뷰 개수")
plt.xlabel("카테고리")
plt.ylabel("수")
plt.tight_layout()
plt.show()
```

## 카테고리별 평균 평점

```python
# 카테고리별 전처리 리뷰 데이터 목록
score_avg_file_paths = {"선크림": "./preprocessed_data/preprocessed_suncream_data.csv",
              "선스틱": "./preprocessed_data/preprocessed_sunstic_data.csv",
              "선쿠션": "./preprocessed_data/preprocessed_suncushion_data.csv",
              "선패치": "./preprocessed_data/preprocessed_sunpatch_data.csv",
              "선스프레이": "./preprocessed_data/preprocessed_sunspray_data.csv"}
```

```python
review_avg = {}
```

```python
# 카테고리별 평균 평점구하기
for category, file_path in score_avg_file_paths.items():
    df = pd.read_csv(file_path)
    review_avg[category] = df['star'].mean()
```

```python
products = list(review_avg.keys())
averages = list(review_avg.values())


plt.figure(figsize=(10, 6))
plt.plot(products, averages, marker='o', color='skyblue')
plt.title('선제품 카테고리별 평균 평점')
plt.xlabel('제품')
plt.ylabel('평균 평점')
plt.ylim(4.6, 4.8)
plt.grid(True)
for i, txt in enumerate(averages):
    plt.text(products[i], averages[i] + 0.005, f'{txt:.2f}', ha='center')


plt.show()
```
